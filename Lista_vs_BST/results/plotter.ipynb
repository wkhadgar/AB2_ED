-------------------
Comparador Busca em Lista x Árvore de Busca Binária
-------------------

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
import pandas as pd

N_MAX = 50
FIXED_MAX = 2000
DYN_MAX = 100
```

```python
# Gera as linhas de base para a comparação.
X_t = np.linspace(1, N_MAX, N_MAX) # Eixo x.
On_linear = 1*(X_t) # Eixo y base da busca linear.
On_BST = np.log(X_t) # Eixo y base da busca binaria.
```

```python
fig, ax = plt.subplots()
plt.plot(X_t, On_BST, color="green")
plt.plot(X_t, On_linear, color="red")
```
